```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
import earthpy.plot as ep


def load(path):
    image = cv2.imread(path)
    #image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    print(image.shape)
    image = cv2.resize(image, (1440, 1120))
    return image/255

def rgb_colors(image):
    plt.imshow(image),plt.show()
    (B_ref, G_ref, R_ref) = cv2.split(image)
    R_ref = np.clip(R_ref, 0, 1)
    G_ref = np.clip(G_ref, 0, 1)
    B_ref = np.clip(B_ref, 0, 1)
    gamma = 2.2
    R = np.power(R_ref, 1/gamma)
    G = np.power(G_ref, 1/gamma)
    B = np.power(B_ref, 1/gamma)
    #np.warnings.filterwarnings('ignore') # Removes warning that array has NaN pixels  
    hist1 = ep.hist(R,
            figsize=(4, 2),
            colors=['red'],
            title=["Distribution of Red Pixels"])
    
    hist2 = ep.hist(G,
            figsize=(4, 2),
            colors=['green'],
            title=["Distribution of Green (NIR Veggie) Pixels"])
    
    hist3 = ep.hist(B,
            figsize=(4, 2),
            colors=['blue'],
            title=["Distribution of Blue Pixels"])


    fig, ([ax1, ax2, ax3]) = plt.subplots(1, 3, figsize=(16, 3))

    ax1.imshow(R, cmap='Reds', vmax=1, vmin=0)
    ax1.set_title('Red', fontweight='bold')
    ax1.axis('off')
    
    ax2.imshow(G, cmap='Greens', vmax=1, vmin=0)
    ax2.set_title('NIR "Veggie"', fontweight='bold')
    ax2.axis('off')
    
    ax3.imshow(B, cmap='Blues', vmax=1, vmin=0)
    ax3.set_title('Blue', fontweight='bold')
    ax3.axis('off')
    
    plt.subplots_adjust(wspace=.02)
```

```python
rgb_colors(load('0724imgs/original.JPG'))
rgb_colors(load('0724imgs/hhyc.JPG'))
rgb_colors(load('0724imgs/hhcy.JPG'))
```
